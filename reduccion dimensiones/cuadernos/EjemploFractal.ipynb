```python
import matplotlib.pyplot as plt
import random
```

```python
def generar_valores_fractal_helecho(iteraciones=50000):
    x = [0]
    y = [0]

    for _ in range(iteraciones):
        r = random.uniform(0, 100)

        if r <= 1:
            xn = 0
            yn = 0.16 * y[-1]
        elif r <= 86:
            xn = 0.85 * x[-1] + 0.04 * y[-1]
            yn = -0.04 * x[-1] + 0.85 * y[-1] + 1.6
        elif r <= 93:
            xn = 0.2 * x[-1] - 0.26 * y[-1]
            yn = 0.23 * x[-1] + 0.22 * y[-1] + 1.6
        else:
            xn = -0.15 * x[-1] + 0.28 * y[-1]
            yn = 0.26 * x[-1] + 0.24 * y[-1] + 0.44

        x.append(xn)
        y.append(yn)

    return x, y
```

```python
def graficar_fractal_helecho(x, y, title):
    plt.scatter(x, y, s=0.2, c='green')
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
```
